# Data Visualization

```python
#import libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load data 
df = pd.read_csv("iris.csv", names = ['sepal_length','sepal_width','petal_length','petal_width','species'])
```

```python
# Display the First Few Rows
print("First Five Rows of the Dataset:")
print(df.head())
```

```python
print("Dataset Information")
print(df.info)
```

```python
# Shape of dataset 
df.shape
```

```python
# Checking for Missing Values 
print("Missing Value")
df.isnull().sum()
```

```python
# Statistical Data 
df.describe()
```

### Visualization 


### a. Countplot

```python
plt.figure(figsize=(5,3))
df['species'].value_counts().plot(kind='bar', color=['#C0C9EE', '#A2AADB', '#898AC4'])
plt.title('Count of Each Iris Species', fontdict={'fontweight':'bold', 'fontsize': 16})
plt.xlabel('SPECIES')
plt.ylabel('COUNT OF SPECIES')
plt.show()
```

### b. Pie Chart

```python
df['species'].value_counts().plot(kind='pie',autopct='%.2f') #autopct is used to show percentage on pie chart
plt.title('Pie Chart', fontdict={'fontweight':'bold', 'fontsize': 16})
plt.show()
```

### c. Histogram

```python
sns.histplot(data = df , x ='sepal_width',hue='species',bins=20,palette={'Iris-setosa': '#BBFBFF', 'Iris-versicolor': '#8DD8FF', 'Iris-virginica': '#4E71FF'})
plt.title('Histogram of Sepal Width by Species',fontdict={'fontweight':'bold', 'fontsize': 16})
plt.xlabel('Sepal Width')
plt.ylabel('Frequency')
```

### d. Scatter Plot

```python
sns.scatterplot(data = df , x= 'petal_length',y='petal_width', hue = 'species')
plt.title('Petal Length vs Petal Width- Scatter Plot',fontdict={'fontweight':'bold', 'fontsize': 16})
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.show()
```

### e. Box Plot

```python
sns.boxplot(x='species', y='sepal_length', data = df)
plt.title('Sepal Length Distribution by Species',fontdict={'fontweight':'bold', 'fontsize': 16})
plt.ylabel('Sepal Length')
plt.show()
```

### f. Pair Plot

```python
sns.pairplot(df)
plt.show()
```
